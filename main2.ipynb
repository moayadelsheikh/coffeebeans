```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
customers = pd.read_pickle("cleaned data\customers_001.pkl")
products = pd.read_pickle("cleaned data\products_001.pkl")
orders = pd.read_pickle("cleaned data\orders_001.pkl")
```

```python
orders.drop(columns=['customer_name', 'email', 'country', 'coffee_type', 'roast_type', 'size', 'unit_price', 'sales'], inplace=True)
```

```python
products.head(5)
```

```python
orders = pd.merge(orders, customers[['customer_id', 'customer_name', 'city', 'country', 'loyalty_card']], on='customer_id', how='left')
orders = pd.merge(orders, products[['product_id', 'coffee_type', 'roast_type', 'size', 'unit_price', 'price_per_100g', 'profit']], on='product_id', how='left')
```

```python
orders['sales'] = orders['quantity'] * orders['unit_price']
orders['net_profit'] = orders['quantity'] * orders['profit']
```
